## Preprocessing:

1.Missing value statistics:

```python
# check any missing values?
df.isnull().values.any()
```

```python
# check the total number of missing values
df.isnull().sum().sum()
```

2.Handle missing values:

```python
df_new = df.dropna() # removes NA values
```

```python
df_new = df[df['beds'].notna()] # # extract the data that has non-na values with 'beds'
```

```python
df_new = df[df['beds'].notna() & df['bedrooms'].notna()] # extract the data that has non-na values with both 'beds' and 'bedrooms'
```

noted that df is edited in-place here 

```python
df['beds'].fillna((df['beds'].mean()), inplace=True) # fill the na values data in 'beds' with the mean value and modify any other views on this object
```

3.Categorical encoding:

```python
print(list(df.room_type)[:10]) # printing first-10 rows' room types 
```

encoding

```python
encoded_room_type_columne = df.room_type.astype('category').cat.codes # categorized each of the variables into numbers
print(list(encoded_room_type_columne)[:10]) #printing first-10 rows' encoding
```

```python
df.room_type = encoded_room_type_columne # replace original data with encoding
print(list(df.room_type)[:10]) #printing first-10 rows' room types 
```
